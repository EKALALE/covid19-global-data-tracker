# COVID-19 Global Data Tracker
This notebook provides an overview of global COVID-19 data including confirmed cases, recoveries, and deaths by country.

```
import pandas as pd
import plotly.express as px
```

```
# Sample COVID-19 data
data = {
    'Country': ['USA', 'India', 'Brazil', 'Russia', 'France'],
    'Confirmed': [105000000, 45000000, 38000000, 23000000, 25000000],
    'Recovered': [98000000, 44000000, 37000000, 22000000, 24500000],
    'Deaths': [1200000, 530000, 700000, 400000, 300000]
}
df = pd.DataFrame(data)
df
```

```
# Bar chart for visualization
fig = px.bar(df, x='Country', y=['Confirmed', 'Recovered', 'Deaths'],
             title='COVID-19 Global Cases by Country',
             labels={'value': 'Cases', 'variable': 'Status'},
             barmode='group')
fig.show()
```
